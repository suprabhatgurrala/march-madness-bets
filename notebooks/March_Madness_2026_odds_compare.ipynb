# 2026 March Madness

Notebook to compare Nate Silver's model predictions vs Bovada betting odds.

```python
bankroll = 100
```

```python
import pandas as pd

from march_madness_bets import data, optimizer

pd.set_option("display.max_columns", None)

merged_df = data.get_combined_data(include_alt_spreads=True)

# Filter to only games today
merged_df = merged_df[
    merged_df.event_time.dt.date == merged_df.event_time.dt.date.min()
]
```

```python
merged_df = data.compute_log_ev(merged_df, bankroll=bankroll)
```

```python
# Filter out obviously unprofitable bets
rec_bets = merged_df[merged_df["kelly"] > 0]
rec_bets
```

```python
rec_bets.loc[0].sort_values("log_ev", ascending=False)
```

```python
rec_bets[["bet_name", "odds", "prob_silver"]].style.format(
    {"prob_silver": "{:.2%}"}
).hide(axis="index")
```

```python
optimizer.optimize_kelly(rec_bets.reset_index(), bankroll=bankroll).set_index(
    ["game_id", "event_time", "event_name"]
).sort_index()
```
