# 2026 March Madness

Notebook to compare Nate Silver's model predictions vs Bovada betting odds.

```python
bankroll = 100
```

```python
import pandas as pd

from march_madness_bets import data
from march_madness_bets.optimizer import multi_kelly_binary

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
merged_df = data.get_combined_data(include_alt_spreads=True)

# Filter games that started already
merged_df = merged_df[merged_df["event_time"] > pd.Timestamp.now(tz="US/Pacific")]

# Filter to only games today
merged_df = merged_df[
    merged_df.event_time.dt.date == merged_df.event_time.dt.date.min()
]
```

```python
merged_df = data.compute_log_ev(merged_df, bankroll=bankroll)
```

```python
# Filter out obviously unprofitable bets
rec_bets = merged_df[merged_df["kelly"] > 0]
```

```python
rec_bets["optimal_wager"] = multi_kelly_binary(
    game_ids=rec_bets.game_id.values,
    odds=rec_bets.odds.values,
    probs=rec_bets.prob_silver.values,
    bankroll=bankroll,
)
```
